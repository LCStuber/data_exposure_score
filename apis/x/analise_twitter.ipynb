```python
import tweepy
import os

API_KEY = os.getenv("API_KEY")
API_SECRET = os.getenv("API_SECRET")
ACCESS_TOKEN = os.getenv("ACCESS_TOKEN")
ACCESS_TOKEN_SECRET = os.getenv("ACCESS_TOKEN_SECRET")
BEARER_TOKEN = os.getenv("BEARER_TOKEN")


API_KEY_2 = os.getenv("API_KEY_2")
API_SECRET_2 = os.getenv("API_SECRET_2")
ACCESS_TOKEN_2 = os.getenv("ACCESS_TOKEN_2")
ACCESS_TOKEN_SECRET_2 = os.getenv("ACCESS_TOKEN_SECRET_2")
BEARER_TOKEN_2 = os.getenv("BEARER_TOKEN_2")

client = tweepy.Client(
    bearer_token=BEARER_TOKEN_2,
    consumer_key=API_KEY_2,
    consumer_secret=API_SECRET_2,
    access_token=ACCESS_TOKEN_2,
    access_token_secret=ACCESS_TOKEN_SECRET_2,
    wait_on_rate_limit=True
)
```

```python
username = "MatCMartins"

try:
    # Buscar detalhes do usu√°rio
    user = client.get_user(username=username)
    
    # Verificar se o usu√°rio foi encontrado
    if user.data:
        print(f"User ID: {user.data.id}")
        print(f"Nome: {user.data.name}")
        print(f"Username: @{user.data.username}")
    else:
        print("Usu√°rio n√£o encontrado.")

except tweepy.TweepyException as e:
    print(f"Erro ao buscar usu√°rio: {e}")
```

```python
# ID do usu√°rio (@MatCMartins)
user_id = 1549737028643917825

import csv

try:
    tweets = client.get_users_tweets(id=user_id, max_results=100, tweet_fields=["created_at", "text"])

    if tweets.data:
        with open("tweets.csv", mode="w", newline="", encoding="utf-8") as file:
            writer = csv.writer(file)
            writer.writerow(["Tweet ID", "Data", "Texto"])

            for tweet in tweets.data:
                writer.writerow([tweet.id, tweet.created_at, tweet.text])

                print(f"Tweet ID: {tweet.id}")
                print(f"Data: {tweet.created_at}")
                print(f"Texto: {tweet.text}")
                print("-" * 50)
    else:
        print("Nenhum tweet encontrado para este usu√°rio.")

except tweepy.TweepyException as e:
    print(f"Erro ao buscar tweets: {e}")
```
